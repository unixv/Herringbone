```python
import socket

def send_udp_message(message, server_ip, server_port):
    # Create a UDP socket
    udp_client_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    try:
        # Send message to the server
        udp_client_socket.sendto(message.encode(), (server_ip, server_port))
        print(f"Message sent to {server_ip}:{server_port}")
    finally:
        # Close the socket
        udp_client_socket.close()

if __name__ == "__main__":
    server_ip = "127.0.0.1"  # Replace with the server's IP address
    server_port = 7000      # Replace with the server's port number
    message = "{\"body\":\"body\"}""

    send_udp_message(message, server_ip, server_port)
```
